```python
from sklearn.preprocessing import StandardScaler
feature = df.drop(['총_매출액'], axis=1).columns

x = df.drop(['총_매출액'], axis=1).values
y = df['총_매출액'].values
x = StandardScaler().fit_transform(x)
x = pd.DataFrame(x, columns=feature)
x
```

1) 주성분분석은 스케일을 맞춰야함
2) 데이터프레임(df)에서 '총_매출액' 열을 제외한 열들의 이름을 추출하여 feature 변수에 저장합니다.
3) (axis=1   #1은변수 제거 0은 행제거)
3,4) '총_매출액' 열을 제외한 열들의 데이터를 x에 저장하고, '총_매출액' 열의 데이터를 y에 저장합니다.

5) StandardScaler를 사용하여 특성들을 표준화합니다. 표준화는 각 특성의 평균을 0으로, 표준편차를 1로 만들어 데이터의 단위와 크기를 맞추는 작업입니다. fit_transform 메서드를 사용하여 데이터를 표준화한 후, 이를 변수 x에 할당합니다.
6) 표준화된 데이터를 다시 데이터프레임 형태로 변환하고, 컬럼 이름을 feature에 저장된 열 이름으로 설정합니다.
이 코드는 주어진 데이터프레임에서 '총_매출액' 열을 제외한 열들을 표준화한 후, 다시 데이터프레임으로 변환하는 과정을 나타내고 있습니다.

1. random성을 고정시켜주는 파라미터다.

```python
import matplotlib.pyplot as plt

percent_variance = np.round(pca.explained_variance_ratio_*100, 2)  #퍼센트로 나타나기 위해 *100
columns = []
for i in range(len(percent_variance)):
  columns.append(f'PC{i+1}')
fig = plt.figure()  #도화지 만들기
fig.set_size_inches(20,10) #도화지 사이즈
ax = plt.bar(x = range(len(percent_variance)), height=percent_variance, tick_label=columns)
plt.ylabel('Percentate of Variance Explaiend')
plt.xlabel('Principal Component')
plt.show
```
